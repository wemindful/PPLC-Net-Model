```python
import random
import cv2
from matplotlib import pyplot as plt
import albumentations as A

# 参考文档
# https://albumentations.ai/docs/api_reference/augmentations/transforms/#albumentations.augmentations.transforms.RandomToneCurve

def visualize(image):
    plt.figure(figsize=(10, 5))
    plt.axis('off')
    plt.imshow(image)
```

```python
image = cv2.imread('D:\\0019_0003.JPG')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
visualize(image)
# cv2.imwrite("D://ss.jpg",img=cv2.cvtColor(image, cv2.COLOR_RGB2BGR))
```

```python
transform = A.Compose(
    [A.HorizontalFlip(p=0.5)],
)
random.seed(7)
transformed = transform(image=image)
visualize(transformed['image'])
cv2.imwrite("D://Rain.jpg",img=cv2.cvtColor(transformed['image'], cv2.COLOR_RGB2BGR))
```

```python
transform = A.Compose(
    #[A.Blur((3, 5),p=0.5)], 
    [A.MotionBlur((3, 3),True,p=1)],
)
random.seed(7)
transformed = transform(image=image)
visualize(transformed['image'])
cv2.imwrite("D://Rain.jpg",img=cv2.cvtColor(transformed['image'], cv2.COLOR_RGB2BGR))
```

```python
transform = A.Compose(
    [A.MedianBlur(p=0.5)], 
)
random.seed(7)
transformed = transform(image=image)
visualize(transformed['image'])
cv2.imwrite("D://Rain.jpg",img=cv2.cvtColor(transformed['image'], cv2.COLOR_RGB2BGR))
```

```python
transform = A.Compose(
    [A.RandomBrightnessContrast (p=1)],  # 随机改变输入图像的亮度和对比度
)
random.seed(7)
transformed = transform(image=image)
visualize(transformed['image'])
cv2.imwrite("D://Rain.jpg",img=cv2.cvtColor(transformed['image'], cv2.COLOR_RGB2BGR))
```

```python
transform = A.Compose(
    [A.MultiplicativeNoise(multiplier=1.5,p=1)], # 图像乘以随机数或数字数组
)
random.seed(7)
transformed = transform(image=image)
visualize(transformed['image'])
cv2.imwrite("D://Rain.jpg",img=cv2.cvtColor(transformed['image'], cv2.COLOR_RGB2BGR))
```

```python
transform = A.Compose(
    [A.Downscale(scale_min=0.9,scale_max=0.9,always_apply=True,p=1)],  # 信号干扰降低图像质量
)
random.seed(7)
transformed = transform(image=image)
visualize(transformed['image'])
cv2.imwrite("D://Rain.jpg",img=cv2.cvtColor(transformed['image'], cv2.COLOR_RGB2BGR))
```

```python
transform = A.Compose(
    [A.PixelDropout(dropout_prob=0.45,per_channel=False,drop_value=3,p=1)],  # 信号干扰降低图像质量 以一定的概率将像素设置为 0。
)
random.seed(7)
transformed = transform(image=image)
visualize(transformed['image'])
cv2.imwrite("D://Rain.jpg",img=cv2.cvtColor(transformed['image'], cv2.COLOR_RGB2BGR))
```

```python
transform = A.Compose(
    [A.Spatter(mean=0.68,p=1,mode="mud")],  # 应用飞溅变换。它模拟了可能以雨或泥的形式遮挡镜头的损坏。
)
random.seed(7)
transformed = transform(image=image)
visualize(transformed['image'])
cv2.imwrite("D://Rain.jpg",img=cv2.cvtColor(transformed['image'], cv2.COLOR_RGB2BGR))
```

```python
transform = A.Compose(
     [A.RandomSunFlare(flare_roi=(0, 0, 0.1, 0.5), angle_lower=0.0,angle_upper=1,num_flare_circles_lower=1,num_flare_circles_upper=2, p=1)],
   #[A.RandomSunFlare(flare_roi=(0.9, 0, 1, 0.5), angle_lower=0.5, p=1)],
)
#random.seed(3)
transformed = transform(image=image)
visualize(transformed['image'])
cv2.imwrite("D://SunFlare.jpg",img=cv2.cvtColor(transformed['image'], cv2.COLOR_RGB2BGR))
```

```python
transform = A.Compose(
    [A.RandomShadow(num_shadows_lower=1, num_shadows_upper=5, shadow_dimension=6, shadow_roi=(0, 0.5, 1, 1), p=1)],
)
random.seed(3)
transformed = transform(image=image)
visualize(transformed['image'])
cv2.imwrite("D://Shadow.jpg",img=cv2.cvtColor(transformed['image'], cv2.COLOR_RGB2BGR))
```

```python
transform = A.Compose(
    [A.RandomFog(fog_coef_lower=0.1, fog_coef_upper=0.2, alpha_coef=0.2, p=1)],
)
random.seed(7)
transformed = transform(image=image)
visualize(transformed['image'])
cv2.imwrite("D://Fog.jpg",img=cv2.cvtColor(transformed['image'], cv2.COLOR_RGB2BGR))
```
